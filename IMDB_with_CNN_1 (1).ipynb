```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.datasets import imdb
from tensorflow import keras
from tensorflow.keras import datasets, regularizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.layers import Embedding
from tensorflow.keras.layers import Dense,Dropout,BatchNormalization, Conv1D, GlobalMaxPooling1D, Flatten,MaxPooling1D
import seaborn as sns

```

#Loading Data , Preprocessing and Normalization

```python
top_words = 10000
input_length = 1000


(x_train, y_train), (x_test, y_test) = datasets.imdb.load_data(num_words=top_words)

x_train = pad_sequences(x_train, maxlen=input_length)
x_test = pad_sequences(x_test, maxlen=input_length)

from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size = 0.20, random_state = 42, stratify=y_train)

from sklearn import preprocessing

x_scaled = preprocessing.normalize(x_train)
df = pd.DataFrame(x_scaled) 
x_scaled_test = preprocessing.normalize(x_train)
df_test = pd.DataFrame(x_scaled_test) 


```

```python
from sklearn import preprocessing

x_scaled = preprocessing.normalize(x_train)
df = pd.DataFrame(x_scaled) 
x_scaled_test = preprocessing.normalize(x_test)
df_test = pd.DataFrame(x_scaled_test) 

import seaborn as sns
import matplotlib.pyplot as plt

review_len_train = []
review_len_test = []
for i,j in zip(x_scaled,x_scaled_test ):
    review_len_train.append(np.mean(i))
    review_len_test.append(np.mean(j))

sns.distplot(review_len_train,hist_kws={"alpha":0.3});
```

#The Model

```python
optimizers = ['adam','sgd']
dropoutRates = [0.2,0.3,0.4]
for optimizer in optimizers:
    for dropout in dropoutRates:
        embedding_size = 32
        model = Sequential()
        model.add(Embedding(top_words, embedding_size, input_length=input_length))
        model.add(Conv1D(filters=64, kernel_size=3,strides=1, padding='same', activation='relu'))
        model.add(MaxPooling1D(pool_size=5))
        model.add(GlobalMaxPooling1D())
        model.add(Flatten())
        model.add(Dense(250, activation='relu'))
        model.add(Dropout(dropout))
        model.add(Dense(1, activation='sigmoid'))


        model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
        print(model.summary())
        print("optimizer: ",optimizer,", dropout: ",dropout)
        #callback = tf.keras.callbacks.EarlyStopping(patience=5)

        #batch_size16
        history = model.fit(x_train, y_train, batch_size=128, validation_data=(x_val, y_val), validation_batch_size=128 , epochs=5, verbose = 1)

        loss, accuracy = model.evaluate(x_test, y_test, verbose=2)
        print('Loss : ', loss)
        print('Acc :', accuracy*100)

        plt.plot(history.history['loss'])
        plt.plot(history.history['val_loss'])
        plt.title('Model loss')
        plt.xlabel('epoch')
        plt.ylabel('loss')
        plt.legend(['train_lose', 'val_loss'])
        plt.show()
```

```python
embedding_size = 32
model = Sequential()
model.add(Embedding(top_words, embedding_size, input_length=input_length))
model.add(Conv1D(filters=64, kernel_size=3,strides=1, padding='same', activation='relu'))
model.add(MaxPooling1D(pool_size=5))
model.add(GlobalMaxPooling1D())
model.add(Flatten())
model.add(Dense(250, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))


model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
print("optimizer: ",'adam',", dropout: ",'0.3')
#callback = tf.keras.callbacks.EarlyStopping(patience=5)

#batch_size16
history = model.fit(x_train, y_train, batch_size=128, validation_data=(x_val, y_val), validation_batch_size=128 , epochs=5, verbose = 1)

loss, accuracy = model.evaluate(x_test, y_test, verbose=2)
print('Loss : ', loss)
print('Acc :', accuracy*100)

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train_lose', 'val_loss'])
plt.show()
```

#Confusion Matrix


```python
from sklearn.metrics import confusion_matrix
#
# Get the predictions
#
y_pred = model.predict(x_test)
y_pred = np.where(y_pred > 0.5, 1, 0)
#
# Calculate the confusion matrix
#
conf_matrix = confusion_matrix(y_test, y_pred)
#
# Print the confusion matrix using Matplotlib
#
fig, ax = plt.subplots(figsize=(7.5, 7.5))
ax.matshow(conf_matrix, cmap=plt.cm.Blues, alpha=0.3)
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='xx-large')
 
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
print(len(x_test))
print(len(y_pred))
```

#ROC Curve


```python
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

def plot_roc_curve(true_y, y_prob):
    """
    plots the roc curve based of the probabilities
    """

    fpr, tpr, thresholds = roc_curve(true_y, y_prob)
    plt.plot(fpr, tpr)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')

plot_roc_curve(y_test, y_pred)
print(f'model 1 AUC score: {roc_auc_score(y_test, y_pred)}')
```
